```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv("extracted_features_charrec.csv")

# Define the initial weights and learning rate
W = np.array([10, 0.2, -0.75])
learning_rate = 0.05

# Define activation functions
def bipolar_step_activation(x):
    return -1 if x < 0 else 1

def sigmoid_activation(x):
    return 1 / (1 + np.exp(-x))

def relu_activation(x):
    return max(0, x)

def step_function(x):
  if x>0:
    return 1
  else:
    return 0

def summation_unit(x1,x2,w0,w1,w2):
  return (w0)+(x1*w1)+(x2*w2)

def summation_data(x0,x1,x2,w0,w1,w2):
  return (x0*w0)+(x1*w1)+(x2*w2)

def perceptron_train(inputs, targets, weights, learning_rate, max_epochs, convergence_error, activation_func):
    error_values = []  # To store error values for plotting

    for epoch in range(max_epochs):
        error_sum = 0

        for i in range(len(inputs)):
            input_data = inputs[i]
            target = targets[i]

            # Compute the predicted output with the specified activation function
            prediction = activation_func(np.dot(input_data, weights))

            # Update weights
            error = target - prediction
            weights += learning_rate * error * input_data

            # Calculate error for this sample
            error_sum += error ** 2

        # Append the error to the list for plotting
        error_values.append(error_sum)

        # Check for convergence
        if error_sum <= convergence_error:
            break

    return error_values, epoch + 1  # Return error values and the number of epochs
inputs = data[['3', '4', '5']].values
targets = data['class_name'].values


# Set maximum epochs and convergence error
max_epochs = 1000
convergence_error = 0.002

# Perform experiments with different activation functions
activation_functions = [bipolar_step_activation, sigmoid_activation, relu_activation]
activation_names = ['Bi-Polar Step', 'Sigmoid', 'ReLU']
iterations_to_converge = []

for i, activation_func in enumerate(activation_functions):
    W = np.array([10, 0.2, -0.75])  # Reset weights for each activation function
    error_values, epochs = perceptron_train(inputs, targets, W, learning_rate, max_epochs, convergence_error, activation_func)
    iterations_to_converge.append(epochs)
    print(f"{activation_names[i]} Activation:")
    print(f"Number of epochs needed for convergence: {epochs}")
    print(f"Final Error: {error_values[-1]}\n")

# Compare iterations taken to converge
plt.bar(activation_names, iterations_to_converge)
plt.xlabel('Activation Function')
plt.ylabel('Iterations to Converge')
plt.title('Iterations to Converge for Different Activation Functions')
plt.show()
```
